```python
    plt.figure(figsize=(16,5))
    sns.boxplot(x='Sampling Method', y='Measurement', data=df)
    plt.title('Box plots of measurements by sampling method')
    plt.show()
```

Creating a for loop to create individual boxplots, so the scale of one doesn't effect the others.

```python
methods = df['Sampling Method'].unique()
'''
Used ChatGPT, to correct the sns.boxplot portion. was missing y and didnt need the dataframe mask
Orginally only had ...X = "sampling data", data = df[['sampling method', 'Measurement']][df['sampling method'] == method]...
'''
for method in methods:
    plt.figure(figsize=(12,5))
    sns.boxplot(x='Sampling Method', y='Measurement', data=df[df['Sampling Method'] == method])
    plt.title('Box plots of measurements by sampling method')
    plt.show()
```
